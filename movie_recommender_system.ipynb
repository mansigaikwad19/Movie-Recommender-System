this function wont work because of the following error.

for this we will use the ast module of python

```python
import ast
```

```python
def convert(obj):
  name=[]
  for i in ast.literal_eval(obj):
    name.append(i['name'])
  return name
```

```python
df['genres']=df['genres'].apply(convert)
```

We will take first 3 dictionaries and get only the name not the character

```python
def convert_3(obj):
  name=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter != 3:
      name.append(i['name'])
      counter+=1
    else:
      break
  return name
```

```python
df['cast']=df['cast'].apply(convert_3)
```

```python
df.head(2)
```

```python
df['crew'][0]
```

we only want that job where job is director and the name of that director

```python
def fetch_director(obj):
  name=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      name.append(i['name'])
      break
  return name
```

```python
df['crew']=df['crew'].apply(fetch_director)
```

## Stemming

Activity and activities are considered different here.
Here we will apply stemming to solve this problem


```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
pt=PorterStemmer()
```

```python
pt.stem('loving')
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(pt.stem(i))
  return " ".join(y)
```

```python
new_df['tags'][0]
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

Whenever you are working with high dimensionl data, never use eucledian distance . Higher dimensional the data , euclidean distance fails. It is not a good measure.
Use Cosine distance .
Distance is inversely proportional to similarity. 
Smaller the distance greater is the similarity, and greater the distance, lesser the similarity.
Here we have to find the cosine distance of each movie with the other movies.

```python
#cosine similarity function 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

It is an array of array (4806 x 4806)

```python
similarity[0]
```

Here it is giving us the similarity of first movie with the other 4806 movies . including itself.
That is why the first value is 1 (similarity of 1st movie with 1st movie(itself))

```python
similarity[1]
```

```python
sorted(similarity[0],reverse=True)
```

the problem here is after srting we lose the index ie. we would loose the info that which movie has highest similarity with which movie.
Hence use Enumerate function

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
#here we are getting the index and the distance 
#key=lambda x:x[1] sorts based on the similarity not the index number
```

```python
#to get the top 5 similar movies
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

Create a function where if someone gives us a movie name , this RS will give it 5 similar movies

1. USer will give a movie name 
2. get the index of that movie from your data .
3. Index will help get into the similarity metrix.
for eg movie - avatar. 
Index is 0 . then will check the similarith of avatar with each movie.
4. Sort the movies. So that you will get the ones with less distance first and more distance ie less similar once in the last.
5. Recommend the first 5 movies (top 5)

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  
  for i in movies_list:
    print(new_df.iloc[i[0]].title)

#we will find the index here
```

```python
recommend('Batman Begins') 
```
